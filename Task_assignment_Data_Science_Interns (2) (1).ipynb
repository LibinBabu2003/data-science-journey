```python
import pandas as pd
import re

modifier_columns = [
    'movement_sprint_speed',
    'skill_ball_control',
    'power_stamina',
    'attacking_finishing'
]

def convert_to_number(value):
    if isinstance(value, str) and re.match(r"^\d+[+-]\d+$", value):
        base, modifier = re.split(r'[+-]', value)
        operator = '+' if '+' in value else '-'
        return int(base) + int(modifier) if operator == '+' else int(base) - int(modifier)
    elif isinstance(value, str) and value.isdigit():
        return int(value)
    return value

for col in modifier_columns:
    if col in df_main.columns:
        df_main[col] = df_main[col].apply(convert_to_number)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import random
import matplotlib.cm as cm




def scatter(a, b):
    if a in df_main.columns and b in df_main.columns:
        color = random.choice(plt.cm.tab10.colors)
        sns.scatterplot(data=df_main, x=a, y=b,color=color)
        plt.title(f"{a} vs {b}")
        plt.xlabel(f"{a}")
        plt.ylabel(f"{b}")
        plt.xticks(rotation=90)
        plt.tight_layout()
        plt.show()
    else:
        print(f" Skipping: '{a}' or '{b}' not found in DataFrame.")


def boxplot(a,x, y):
    if x in a.columns and y in a.columns:

        sns.boxplot(data=a, x=x, y=y)
        plt.title(f"{x} vs {y} (Box Plot)")
        plt.xlabel(f"{x}")
        plt.ylabel(f"{y}")
        plt.xticks(rotation=90)
        plt.tight_layout()
        plt.show()
    else:
        print(f" Skipping: '{x}' or '{y}' not found in DataFrame.")



def barplot(a,x, y):
    if x in a.columns and y in a.columns:
        categories = sorted(a[x].dropna().unique())
        colors = random.sample(list(sns.color_palette("hsv", len(categories))), len(categories))
        palette_map = dict(zip(categories, colors))
        sns.barplot(data=a, x=x, y=y, palette=palette_map, hue=x, legend=False)
        plt.title(f"{x} vs {y} (Bar Plot)")
        plt.xlabel(f"{x}")
        plt.ylabel(f"{y}")

        plt.xticks(rotation=90)
        plt.tight_layout()
        plt.show()
    else:
        print(f" Skipping: '{x}' or '{y}' not found in DataFrame.")

def lineplot(a,x,y):
  if x in a.columns and y in a.columns:
        color = random.choice(plt.cm.tab10.colors)
        sns.lineplot(data=a, x=x, y=y,markers='o',color=color)
        plt.title(f"{x} vs {y} (Line Plot)")
        plt.xlabel(f"{x}")
        plt.ylabel(f"{y}")

        plt.xticks(rotation=90)
        plt.tight_layout()
        plt.show()
  else:
    print(f" Skipping: '{x}' or '{y}' not found in DataFrame.")




scatter("age", "pace")
lineplot(df_main,"age","pace")
```

```python
def plotsca(data, x, y, hue=None, title=None, rotation=90,figsize=(10,6), palette='hsv', legend=True):
    plt.figure(figsize=figsize)
    sns.scatterplot(data=data, x=x, y=y, hue=x, palette=palette)
    plt.title(title or f"{x} vs {y}")
    plt.xlabel(x)
    plt.ylabel(y)
    plt.xticks(rotation=rotation)
    plt.legend(title=y, bbox_to_anchor=(1.05, 1), loc='upper left')
    plt.tight_layout()
    plt.show()
```

```python
club_wage_total = df_main.groupby('club')['wage_eur'].sum().reset_index()
club_wage_total = club_wage_total.sort_values('wage_eur', ascending=False)
top_Club=club_wage_total.head(20)
lineplot(top_Club, "club", "wage_eur")
plotsca(top_Club, "club", "wage_eur")
```

Top-paying clubs include Real Madrid, Juventus, and Arsenal, each with wages exceeding €14 million. English Premier League clubs dominate the top half, indicating high wage investment.

```python
top_player=df_main[['short_name','value_eur']].sort_values(by='value_eur',ascending=False).head(20)

plotsca(top_player, "short_name", "value_eur")
top_player_wage=df_main[['short_name','wage_eur']].sort_values(by='wage_eur',ascending=False).head(20)

plotsca(top_player_wage, "short_name", "wage_eur")
```

COMPARING WAGE_EUR VS VALUE_EUR OF TOP PLAYERS   Most players in both plots overlap, which means high value often corresponds to high wages.However, some veteran players (like Sergio Ramos, Modrić, David Silva) have high wages but lower market values, likely due to age.A few younger or underrated players (like M. Reus) may have high value but comparatively lower wages.

**Find out Most valuable player in every position**

```python
df_main['player_positions'] = df_main['player_positions'].fillna('').astype(str)
df_main['player_positions']=df_main['player_positions'].str.split(',')
expand=df_main.explode('player_positions')
expand['player_positions']=expand['player_positions'].apply(lambda x: x.strip())
valuable_play=expand.loc[expand.groupby('player_positions')['value_eur'].idxmax()].drop_duplicates(subset=['player_positions']).sort_values(by='value_eur', ascending=False)[['short_name', 'player_positions', 'value_eur']].reset_index()
print(valuable_play)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 6))
sns.set(style="whitegrid")


sns.barplot(
    data=valuable_play,
    x='player_positions',
    y='value_eur',
    hue='short_name',
    dodge=False,
    palette='mako'
)

for index, row in valuable_play.iterrows():
    plt.text(index, row['value_eur'] + 1e5, f"{row['short_name']}", ha='center', fontsize=7, rotation=90)
plt.title("Most Valuable Player in Each Position", fontsize=14)
plt.xlabel("Player Position")
plt.ylabel("Market Value (€)")
plt.xticks(rotation=45)
plt.tight_layout()
plt.legend(title='Player', bbox_to_anchor=(1.05, 1), loc='upper left')

plt.show()
```
