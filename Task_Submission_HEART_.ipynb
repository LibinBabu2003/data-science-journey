## Domain analysis

 Domain Analysis Table

| Column               | Description                                                                                                                                   | Data Type  | Expected Range / Notes                                                                 |
|----------------------|-----------------------------------------------------------------------------------------------------------------------------------------------|------------|-----------------------------------------------------------------------------------------|
| `age`                | Age of the patient in completed years. Age is a strong risk factor in heart disease. Older patients typically have higher risk levels.        | int64      | 29 – 77 years                                                                          |
| `sex`                | Sex of the patient. Encoded as 1 for male and 0 for female. Sex can influence heart disease risk and symptoms.                               | int64      | 0 = Female, 1 = Male                                                                   |
| `chest_pain_type`    | Type of chest pain experienced by the patient. Different types indicate varying levels of risk: typical angina, atypical angina, non-anginal, and asymptomatic. | int64 | 0: Typical Angina, 1: Atypical Angina, 2: Non-anginal Pain, 3: Asymptomatic           |
| `resting_bp`         | Resting blood pressure (in mm Hg) measured after the patient has rested. High resting BP is a known cardiovascular risk factor.              | int64      | Typically 90 – 140 mm Hg; values >140 may indicate hypertension                        |
| `cholestoral`        | Serum cholesterol in mg/dL. High levels can indicate an increased risk of coronary artery disease. Includes LDL, HDL, and triglycerides.     | int64      | Normal: < 200 mg/dL; High: > 240 mg/dL                                                 |
| `fasting_blood_sugar`| Indicates whether the fasting blood sugar level is greater than 120 mg/dL. A high value may be a sign of diabetes, which raises heart risks. | int64      | 0 = False, 1 = True                                                                    |
| `restecg`            | Resting electrocardiographic results. Reflects electrical activity of the heart. Certain abnormalities suggest risk of heart disease.        | int64      | 0: Normal, 1: ST-T Wave Abnormality, 2: Left Ventricular Hypertrophy                   |

# Outliers

Box Plot Visualization

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.boxplot(x=df['resting_bp'])
plt.title(f"Boxplot of {'resting_bp'}")
plt.show()

```

 IQR Method (Interquartile Range)

```python
Q1 = df['resting_bp'].quantile(0.25)
Q3 = df['resting_bp'].quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

outliers = df[(df['resting_bp'] < lower_bound) | (df['resting_bp'] > upper_bound)]
print(outliers)
```

```python
print(f"Number of outliers in {'resting_bp'}: {outliers.shape[0]}")
```

 Z-Score Method

```python
from scipy import stats

z_scores = stats.zscore(df['resting_bp'])

threshold = 3

outliers_resting_bp = df[np.abs(z_scores) > threshold]

print(outliers_resting_bp)
print(f"\nNumber of outliers in {'resting_bp'}:{outliers_resting_bp.shape[0]}")
```

**Histogram**

```python
df['resting_bp'].hist()
```

Most values are concentrated around 120 to 140.

This suggests that the majority of patients have normal to slightly elevated blood pressure.

This is expected, as 120/80 mmHg is considered normal, and 130 to 139 is elevated or stage 1 hypertension.

The distribution is right-skewed.

There are fewer patients with very high blood pressure above 160.

A small number of people may have significantly high readings between 180 and 200, which could be outliers.

The distribution appears roughly unimodal.

There is one main peak, or mode, around 130, which might indicate the central tendency of the population.

```python
plt.figure(figsize=(10, 6))
df['resting_bp'].value_counts().plot(kind='bar')
plt.title('resting_bp Distribution')
plt.xlabel('resting_bp')
plt.ylabel('Count')
plt.show()
```

The chart shows a sharp drop as blood pressure rises beyond 140.

Very few patients have resting BP above 160 mmHg.

Some values, like 200 mmHg, appear but are very rare.

**Linegraph**

```python
df_sorted = df.sort_values(by='age')

plt.plot(df_sorted['age'], df_sorted['resting_bp'], label='resting_bp')
plt.xlabel('age')
plt.ylabel('resting_bp')
plt.legend()
plt.show()
```

High Variability:

There is no smooth increase or decrease in the trend.

Blood pressure values vary a lot across different ages.

Slight Increase with Age:

Despite these fluctuations, there is a slight tendency for higher resting BP values in older age groups (50 to 70 years) compared to younger ones.

Outliers & Spikes:

There are several sharp spikes in resting BP, especially between ages 55 and 65. This suggests that some individuals may have unusually high readings or high blood pressure.

No Linear Correlation:

The plot does not show a strong linear link between age and resting BP.

This could mean that age alone does not accurately predict resting BP without taking into account other factors like health status, medications, or lifestyle.
